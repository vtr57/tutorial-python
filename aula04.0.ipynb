# Listas
Listas são uma nova maneira de armazenar dados

## Criando uma lista

```python
dia = [30.2, 55, 20]
```

## Acessando os itens de uma lista

```python
print(f'Primeiro item da lista: {dia[0]}')
print(f'Segundo item da lista: {dia[1]}')
print(f'Terceiro item da lista: {dia[2]}')
```

```python
print(f'Quarto item da lista: {dia[3]}') #erro de índice
```

## Adicionando itens em uma lista

```python
dia.append(72)
print(f'Quarto item da lista: {dia[3]}')
print(dia)
```

## Alterando valor de uma lista

```python
print(dia)
dia[1] = 23
print(f'O valor alterado do segundo item da lista: {dia[1]}')
print(dia)
```
